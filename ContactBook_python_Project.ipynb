```
# defining funcion to welcome user and provide command options
def welcome():
  entry = int(input('''
                      welcome to py contactbook
                      py contactbook commands are : 1,2,3 or 4
                      1.Display your existing contact
                      2.creat a new contact
                      3.check an entry 
                      4.delete an entry
                      5.exit
                      Enter your entry here(1,2,3 or 4):
                      
                      '''
                         ))
  return entry

#defining function phonebook
def phonebook():
  # intialize empty contact dictionary to store contacts
  contact = {}
  # intiate while loop to continously run contact book

  while True:
    entry = welcome()
    # create condition for decision making
    if entry == 1:
      #for first entry check contact list is empty if not print current contact list 
      if bool(contact) != False:
        for k,v in contact.items:
          print(k,'...>',v)
      else :
        print('you have an empty book go to menu add new contact')


    elif entry == 2:
      # request for phone number and contact name
      phone_number = input('please enter a number: ')    
      contact_name = input('what would you like to save as ?')
      if phone_number not in contact.items():
        contact.update({contact_name:phone_number})
        print('print contact successfully saved')
        print('your updated phonebook shown below')
        for k,v in contact.items():
          print(k,'....>',v)
      else :
        print('contact already exists')


    elif entry == 3:
      # intiate new variable user wishes to view 
      name = input('enter the name of contact you wish to view')
      if name in contact:
        print('the contact is ',name,':',contact[name])
      else:
        print('contact does not exist ,return to main menu')


    elif entry == 4:
      #for deleting the contact
      name = input('enter the name of contact you wish to delete:')
      if name in contact:
        print('contact is',name,':',contact[name])
      else :
        print('contact does not  exist return to main menu')
      conform = input('are sure you wish to delete? Yes/No:')

      if conform.capitalize() == 'Yes':
        # if yes delete contact from phone book
        contact.pop(name,None)
        for k,v in contact.items():
          print('your updated phonebook is below')
          print(k,'....>',v)
      else :
        print('return to main menu')
      
    elif entry == 5:
      # break program for fifth entry 
      print('thank you for using py contact book')
      break 

    else :
      # Error message 
      print('Incorrect entry')
```
